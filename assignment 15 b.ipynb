```python
#Q1
'''Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is 
particularly useful for data visualization in scientific computing and data analysis.

Why is Matplotlib Used?
i)Data Visualization: It helps in visualizing data in various formats to make the analysis more intuitive and insightful.
ii)Publication-Quality Figures: It can produce high-quality plots that can be used in academic papers and presentations.
iii)Customizability: It offers extensive customization options for plots, including colors, labels, and styles.'''

#line plot
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()

#bar graph
plt.bar(['A', 'B', 'C', 'D'], [10, 20, 15, 25])
plt.show()

#histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data, bins=4)
plt.show()

#scatter plot
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y)
plt.show()

#pie chart
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels)
plt.show()
```

```python
#Q2
'''A scatter plot is a type of data visualization that displays values for typically two variables for a set of data. The data
is displayed as a collection of points, where each point represents the values of two variables. The position of each point on 
the horizontal and vertical axis indicates values for an individual data point.

Scatter plots are used to observe and show relationships between two numeric variables. If the points are color-coded, one 
additional variable can be displayed. The data points can also be sized to indicate another variable.'''

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y,marker = "v", color = "red")

plt.xlabel("x_value")
plt.ylabel("y_value")
plt.title("scatter plot")
```

```python
#Q3
'''The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to divide the 
plotting area into a grid of subplots and specify which subplot you want to activate for plotting. This is useful for comparing
different plots side by side or displaying different aspects of the same dataset in a coherent and organized manner.'''

import numpy as np
import matplotlib.pyplot as plt

# Data for the plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
plt.figure(figsize=(10, 8))

# Subplot 1
plt.subplot(1, 4, 1)
plt.plot(x1, y1, marker='o')
plt.title('Line Plot 1')
plt.xlabel('X1')
plt.ylabel('Y1')

# Subplot 2
plt.subplot(1, 4, 2)
plt.plot(x2, y2, marker='s')
plt.title('Line Plot 2')
plt.xlabel('X2')
plt.ylabel('Y2')

# Subplot 3
plt.subplot(1, 4, 3)
plt.plot(x3, y3, marker='^')
plt.title('Line Plot 3')
plt.xlabel('X3')
plt.ylabel('Y3')

# Subplot 4
plt.subplot(1, 4, 4)
plt.plot(x4, y4, marker='d')
plt.title('Line Plot 4')
plt.xlabel('X4')
plt.ylabel('Y4')

# Adjust layout
plt.tight_layout()

# Display the plots
plt.show()
```

```python
#Q4
'''A bar plot (or bar chart) is a type of graph that represents categorical data with rectangular bars. Each bar's length or 
height is proportional to the values it represents. Bar plots are commonly used to compare different categories or to track 
changes over time when the changes are relatively large.

Why is it Used?
Comparison: Bar plots are used to compare different categories or groups.
Distribution: They help in visualizing the distribution of data across different categories.
Clarity: They provide a clear and straightforward way to present data.'''

import numpy as np
import matplotlib.pyplot as plt

# Data for the plots
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Creating a vertical bar plot
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.bar(company, profit, color=['blue', 'green', 'red', 'purple'])
plt.title('Company Profits (Vertical)')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')

# Creating a horizontal bar plot
plt.subplot(1, 2, 2)
plt.barh(company, profit, color=['blue', 'green', 'red', 'purple'])
plt.title('Company Profits (Horizontal)')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
#Q5
'''A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of numerical data through
their quartiles. It displays a summary of the data's minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum
values. Box plots are useful for identifying outliers and comparing distributions of data across different groups.

Why is it Used?
Summary of Data: It provides a concise summary of the data distribution, including median and quartiles.
Outlier Detection: It helps in identifying outliers in the data.
Comparison: It allows for easy comparison of distributions between different groups or datasets.'''
import numpy as np
import matplotlib.pyplot as plt

# Generating data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Data')
plt.xlabel('Box')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```
